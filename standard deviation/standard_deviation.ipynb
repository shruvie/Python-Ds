##Visualizing the standard deviation



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statistics as st

d1=[12,34,56,78,90,23,67]

sd=st.stdev(d1)
mean=np.mean(d1)

plt.hist(d1, bins=6,color="orange")

#mean line
plt.axvline(mean,linestyle="--",color="red")

#sd line
plt.axvline(mean-sd,linestyle="-.",color="pink")
plt.axvline(mean+sd,linestyle="-.",color="green")
```
