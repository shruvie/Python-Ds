```python
import matplotlib.pyplot as plt

data = [ [10, 12, 14, 15, 18, 20, 22],
         [8, 9, 11, 13, 17, 19, 21],
         [14, 16, 18, 20, 23, 25, 27] ]

plt.boxplot(data)
plt.xlabel("Groups")
plt.ylabel("Values")
plt.title("Box Plot")
plt.show()
```

#Matplotlib

##Barchart

```python
import matplotlib.pyplot as plt

data=[12,34,5,6]
cat=['a','b','c','d']
plt.bar(cat,data)
plt.xlabel("names")
plt.ylabel("values")
plt.title("raw data")
plt.show()
```

##Line chart


```python
#linechart
import matplotlib.pyplot as plt

data=[2,34,5,67,88,54,22]
plt.plot(data,color='red')
plt.grid('true')
plt.show()
```

##Histogram

```python
##histogram
import matplotlib.pyplot as plt

data=[2,34,5,67,88,54,22]
plt.hist(data,color='red',edgecolor='yellow')
plt.show()
```

##Scatter plot

```python
#scatterplot
import matplotlib.pyplot as plt

data=[23,44,6,678,88,8,65]
y=[2,23,45,56,6,7,7]
plt.scatter(data,y)
plt.show
```
