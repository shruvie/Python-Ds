#Descriptive Statistics

summarizes the statistics, analysis of the whole data.

common libraries used here:
1. scipy
2. numpy
3. pandas

```python
import numpy as np
import pandas as pd
import scipy as sp

a1=[12,23,4,34,56,78]

s=pd.Series(a1)

print("The data: ",a1)
print("The mean: ",s.mean())
print("the median: ",s.median())
print("The mode: ",s.mode())
```

#Descriptive strategy for data analysis

```python
import seaborn as sn
import pandas as pd

#load dataset
df=sn.load_dataset("iris")

#get structure
print("The structure of the dataset \n \nshape: ",df.shape, "\ncolumns: ", df.columns)

#checking misssing values
print("\nSum of missing values", df.isna().sum())

#getting descriptive stats
df.describe(include="all")
```
